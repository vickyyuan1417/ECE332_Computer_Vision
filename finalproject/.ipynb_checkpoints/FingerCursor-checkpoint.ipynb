```python
import cv2
import numpy as np
import mediapipe as mp
from collections import deque
```

```python
def find_hands(image, multi_hand_landmarks):
#     optional in hand marks methodology
    if not multi_hand_landmarks:
        return 0
    imgRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)
    if results.multi_hand_landmarks:
        for hand_pos in results.multi_hand_landmarks: # hand landmarks
            mp.solutions.drawing_utils.draw_landmarks(image, hand_pos, mp.solutions.hands.HAND_CONNECTIONS)  #hand_connections 
    return image

def fingerCursor(dvc):
    my_cam = cv2.VideoCapture(dvc)
    mark = 0
    
#     initialize camera and window
    window_height = my_cam.get(cv2.CAP_PROP_FRAME_HEIGHT)
    window_width = my_cam.get(cv2.CAP_PROP_FRAME_WIDTH)
    once = 1
    twice = 2
    
#     read the gesture image
    gun = cv2.imread('gun.jpg')
    gun = cv2.cvtColor(gun, cv2.COLOR_BGR2GRAY)
    gun, non = cv2.findContours(gun, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)#     read the gun image
    hand_countour_gun = find_hands(gun, mark)#     read the gun image
    
    pointer = cv2.imread('pointer.jpg')
    pointer = cv2.cvtColor(pointer, cv2.COLOR_BGR2GRAY)
    pointer, non = cv2.findContours(pointer, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)#     read the pointer image
    hand_countour_ptr = find_hands(pointer, mark)#     read the pointer image

#     initialization the tip circle position
    pos_finger = (50, 50)
#     initialization the line position
    pos_line = np.append(np.array([], np.uint8), pos_finger)
#     initialization the line length
    length_line = 0
    total_length_line = [0]
#     initialization the filter and kernel
    kernel_ = deque([(50, 50),(50, 50),(50, 50),(50, 50),(50, 50)], 6)
#     initialization the gesture kernel filter
    gesture_kernel = deque([0,0,0,0,0], 6)

    ## color definition
    color1 = (50,0,0)
    color2 = (0,0,255)
    
#     initialization the segmentation kernel
    initial_k = np.ones((10, 10),np.float32)
    initial_k = initial_k / 100

    
# while loop
    while(my_cam.isOpened()):
        non1, sketch = my_cam.read()
        sketch = cv2.flip(sketch, once)
        orig_sk = sketch[:round(window_height),:round(window_width)]

#         skin color check
        bd = cv2.bitwise_and(cv2.cvtColor(orig_sk, cv2.COLOR_BGR2HSV), cv2.cvtColor(orig_sk, cv2.COLOR_BGR2HSV), 
                              mask = cv2.inRange(cv2.cvtColor(orig_sk, cv2.COLOR_BGR2HSV), np.array([0, 30, 45], np.uint8),np.array([50, 245, 255], np.uint8)))
        bd = cv2.erode(bd, initial_k, iterations = once) #set the iteration to 1 while do the countour 
        bd = cv2.dilate(bd, initial_k, iterations = once) #set the iteration to 1 while do the countour 

        colored_img = cv2.cvtColor(bd, cv2.COLOR_HSV2BGR) #rgb info's conbime and output the rgb_img
        cut_img = cv2.cvtColor(colored_img, cv2.COLOR_BGR2GRAY) #gray image
        cut_img = cv2.GaussianBlur(cut_img, (13, 13), 0) #used in lecture 
        
        ct, _non = cv2.findContours(cut_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) 
        
        if not ct:
            print('The fingertip not detected')
            
        if len(ct) != 0:
            cnt_img = max(ct, key = cv2.contourArea)
            if cv2.contourArea(cnt_img) > 800:
                curs_pos = tuple(cnt_img[cnt_img[:,:,1].argmin()][0]) #argmin(0)
                
#                 gun gesture
                num_ges = 0
                my_num = cv2.matchShapes(cnt_img, gun[0], 2, 0.0)   #   gesture should be added here
                gesture_kernel.append(my_num)
            
                def quick_coq(ob):
                    return (ob // 6)
                
                for i in gesture_kernel:
                    num_ges += i
                my_num = quick_coq(num_ges)
                
#                 pointer gesture
                num_ges_pt = 0
                my_num_pt = cv2.matchShapes(cnt_img, pointer[0], 2, 0.0)   #   gesture should be added here
                gesture_kernel.append(my_num_pt)
                
                for i in gesture_kernel:
                    num_ges_pt += i
                my_num_pt = quick_coq(num_ges_pt)
                
                def quick_sf(img):
                    mp.solutions.drawing_utils.draw_landmarks(img, hand_pos, mp.solutions.hands.HAND_CONNECTIONS)
                def quick_dk(ob1, ob2):
                    return (((ob1[0]-ob2[0])**2 + (ob1[1]-ob2[1])**2)**0.5)
                def quick_deq(ob):
                    return (ob // 6)

                length_line = quick_dk(curs_pos, pos_finger)
                if length_line >= 300:
                    pass
                if length_line < 300: 
                    try:
                        cv2.drawContours(colored_img, [cnt_img], 0 , (0, 255, 0), 5)
                        kernel_.append(curs_pos)
                        k_width = 0
                        k_length = 0
                        for i in kernel_:
                            k_length += i[1]
                            k_width += i[0]
                        curs_pos = (quick_deq(k_width), quick_deq(k_length))
                        
                        if not curs_pos:
                            print('cursor not founded, error occurs')

                        if my_num > 0.75: #theshold number set to 0.75
                            pos_line = np.append(pos_line, curs_pos)
                            total_length_line.append(length_line)
                            
                        else:
                            pos_line = np.array([], np.uint8)
#                             the position get updated
                            pos_line = np.append(pos_line, pos_finger) #find the position of pos_line
                            length_line = 0
                            total_length_line = [length_line]
                            pass
#                         the postion updated
                        pos_finger = curs_pos
                    except:
                        pass
            
        for n in range(once, len(total_length_line)):
            cv2.line(orig_sk, (pos_line[n * twice-twice], pos_line[n * twice-once]), (pos_line[n*twice],pos_line[n*twice+once]), color2, round(-0.0328 * total_length_line[n] + 15))
            cv2.line(colored_img, (pos_line[n * twice-twice], pos_line[n * twice-once]), (pos_line[n*twice],pos_line[n*twice+once]), color2, round(-0.0328 * total_length_line[n] + 15))

        cv2.circle(orig_sk, pos_finger, 8, color1 , round(3.024))
        cv2.circle(colored_img, pos_finger, 8, color1 , round(3.024))
        
        cv2.imshow('Finger_cursor_result', sketch)
        if cv2.waitKey(1) & 0xFF == ord('p'): #if press p, the program ends
            my_cam.release()
            break

    my_cam.release()
    cv2.destroyAllWindows()
```

```python
if __name__ == '__main__':
     #the camera device is set to be default to 0
    fingerCursor(0)
```
